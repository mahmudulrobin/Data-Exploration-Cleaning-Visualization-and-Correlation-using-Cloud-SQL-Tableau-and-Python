```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('ggplot')

import matplotlib
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) 
```

```python
# import the dataset
data = pd.read_csv('movies.csv')
data.head(5)
```

```python
# find missing data
for col in data.columns:
    percent_missing = np.mean(data[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
for column in data.columns:
    data.fillna(0, inplace = True)
```

```python
# data type for columns
data.dtypes
```

```python
# Change datatype of columns
data['budget'] = data['budget'].astype('int64')
data['gross'] = data['gross'].astype('int64')
data['runtime'] = data['runtime'].astype('int64')
data['votes'] = data['votes'].astype('int64')
```

```python
data = data.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
pd.options.display.max_rows = 20
```

```python
# delete dublicate rows

data['company'].drop_duplicates()
```

```python
# scatter plot with budget vs gross revenue
plt.scatter(data['budget'], data['gross'])
plt.title('Budget vs Gross Earnings')
plt.ylabel('Gross earnings')
plt.xlabel('Budget')

plt.show()
```

```python
data.head(5)
```

```python
# regression plot using seaborn
sns.regplot('budget', 'gross', data, scatter_kws = {"color" : "blue"}, line_kws = {"color" : "black"})
plt.show()
# these two features are correlated
```

```python
# specific values of correlation between budget and gross income 
data.corr(method = 'pearson') # pearson correlation others kendall, spearman returns little bit different corr values 
```

```python
correlation_matrix = data.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation between gross income and budget')
plt.xlabel('movie features')
plt.ylabel('movie features')


plt.show()
```

```python
# convert company names to numbers
data_neumerized = data.copy(deep = True)

for col_name in data_neumerized.columns:
    if data_neumerized[col_name].dtype == 'object':
        data_neumerized[col_name] = data_neumerized[col_name].astype('category')
        data_neumerized[col_name] = data_neumerized[col_name].cat.codes
    
    
data_neumerized
```

```python
data
```

```python
correlation_matrix = data_neumerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation between gross income and budget')
plt.xlabel('movie features')
plt.ylabel('movie features')


plt.show()
```

```python
# unstacking the correlation matrix
correlation_matrix = data_neumerized.corr()
corr_unstacked = correlation_matrix.unstack()

corr_unstacked
```

```python
sorted_corr_unstacked = corr_unstacked.sort_values()

sorted_corr_unstacked
```

```python
high_correlation = sorted_corr_unstacked[sorted_corr_unstacked > 0.5]

high_correlation
```
